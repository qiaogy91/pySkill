```python
import time
%reset -f
import time
import numpy as np
import matplotlib.pyplot as plt

# 损失函数、以及损失函数的导函数、步长
fx = lambda x: (x - 3.5) ** 2 - 4.5 * x + 10
d = lambda x: 2 * (x - 3.5) - 4.5
step = 0.1

x1 = np.random.randint(0, 12, size=1)[0]  # 随机开始点
lst = []  # 记录X轴上每次迭代的点

while True:
    print(f'------ {x1}')
    
    xnew = x1 - step * d(x1) # 下一次到达的X点
    # 
    # 1.与真实值间的误差
    # if np.abs(xnew - x1) < 0.00001:
    #     break
    #     
    # 2.梯度近乎0
    # if np.abs(d(xnew)) < 0.00001:
    #     break
    #     
    # # 3.损失函数变化近乎0
    # if np.abs(fx(xnew) - fx(x1)) / np.abs(fx(x1)) < 0.00001:
    #     break 
    x1 = xnew
    lst.append(x1)

# 损失函数线条
x = np.linspace(5.75 - 5, 5.75 + 5, 100)
y = fx(x)

# 真实迭代数据
x_ = np.array(lst)
y_ = fx(x_)

# 绘图
axes = plt.subplot(111)
axes.plot(x, y, color='green')
axes.scatter(x_, y_, color='red')
```
