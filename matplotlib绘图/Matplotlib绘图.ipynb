## Matplotlib

### 获取画板

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 创建画板
fig = plt.figure(figsize=(6, 3))  # 宽高（寸）
fig.text(0.5, 1, "images") 				# 画板注释(0, 0) 是画布左下角，(1, 1) 是画布右上角

# 创建画布（必选）
axes1 = plt.subplot(221) # (2,2,1) 将画板分为2行2列，并将画布贴在第一个位置上；
axes2 = plt.subplot(224) # (2,2,4) 将画板分为2行2列，并将画布贴在第四个位置上；

# 画布网格
axes1.grid()
axes2.grid(axis="y", color="red", ls="--")

# 画布注释
axes1.text(2.5,0.5,"qiaogy1") # arg1,arg2 表示将注释添加到坐标中哪个位置
axes2.text(2.5,0.5,"qiaogy2")

# 数据准备（必选）
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 进行绘图（必选）
axes1.plot(x, y1, label='sin(x)', color="green")
axes2.plot(x, y2,label='cos(x)', color="blue")

# 展示Label 图例
axes1.legend(loc='upper left') # 展示图例 
axes2.legend(loc='upper left') # 展示图例 


# 保存图片
fig.savefig("./qioagy.png", dpi=300) # dpi 为像素，像素越大保存的图片越大
```

### 通用属性

```python
# 刻度，一个一个划分的竖线；刻度标签1、2、3
# 默认情况是，画布刻度的边界是由绘图的数据决定的
axes = plt.subplot(2, 1, 1)
axes.plot(x, y1)

# 画布标题、坐标轴标签
axes.set_title('y = F(x)', color='red', fontsize=10)
axes.set_xlabel("X-Label", color='red', fontsize=10)
axes.set_ylabel("Y-Label", color='red', fontsize=10)

# 设置刻度边界
# axes.axis([x.min(),x.max()+2, y1.min(),y1.max()+2])  # [xmin, xmax, ymin, ymax]
axes.set_xlim(0, 10)
axes.set_ylim(-5, 5)


# 设置刻度、刻度标签
axes.set_xticks([0, (1 / 2) * np.pi / 2, np.pi, (3 / 2) * np.pi / 2, 2 * np.pi], ['0', 'Π/2', 'Π', '3Π/2', '2Π'])
axes.set_yticks([-1, 0, 1], ["min", "0", "max"])
```

### 多个画板

```python
# 两张不同的画布
# 一个画板 + 画布
y1 = np.sin(x)
plt.plot(x, y1)
plt.title('SIN(X)')
plt.show()  # 强制当前的所有绘图信息进行渲染（撕下来），渲染完成后会重新生成一个画布进行绘画（之前的就没用了）

# 另一个画板 + 画布
y2 = np.cos(x)
plt.plot(x, y2)
plt.title('COS(X)')
```

### 直方图

【直方图的参数只有一个x！！！不像条形图需要传入x,y】

hist()的参数
+ bins  
可以是一个bin数量的整数值，也可以是表示bin的一个序列。默认值为10
+ normed 现在叫 density

如果值为True，直方图的值将进行归一化处理，形成概率密度，默认值为False
+ color  
指定直方图的颜色。可以是单一颜色值或颜色的序列。如果指定了多个数据集合，颜色序列将会设置为相同的顺序。如果未指定，将会使用一个默认的线条颜色
+ orientation  
通过设置orientation为horizontal创建水平直方图。默认值为vertical

### 饼图

【饼图也只有一个参数x！】

pie()  
饼图适合展示各部分占总体的比例，条形图适合比较各部分的大小

- labels参数设置每一块的标签；
- labeldistance参数设置标签距离圆心的距离（比例值,只能设置一个浮点小数）
- autopct参数设置比例值的显示格式(%1.1f%%)；
- pctdistance参数设置比例值文字距离圆心的距离
- explode参数设置每一块顶点距圆形的长度（比例值,列表）；
- colors参数设置每一块的颜色（列表）；
- shadow参数为布尔值，设置是否绘制阴影
- startangle参数设置饼图起始角度

```python
# 填满的饼图
data = np.array([30, 60, 45])
index = ['A', 'B', 'C']

_ = plt.pie(data, labels=index, labeldistance=1.1, autopct='%1.1f%%',
            pctdistance=0.5, explode=[0, 0, 0.1], colors=np.random.random(size=(3, 3)),
            shadow=True, startangle=10)
_ = plt.axis('equal')
```
